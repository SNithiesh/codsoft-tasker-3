```python
from sklearn.cluster import KMeans
i_rng = range(1,10)
sse=[]

for i in i_rng:
    im = KMeans(n_clusters=i)
    im.fit(df[['petal_length', 'petal_width']])
    sse.append(im.inertia_)
```

```python
sse
```

```python
plt.xlabel('i_rng')
plt.ylabel('Sum of squared errors')
plt.plot(i_rng,sse)
```
